```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
test_d = pd.read_csv('/kaggle/input/playground-series-s5e12/test.csv')
train_d = pd.read_csv('/kaggle/input/playground-series-s5e12/train.csv')
submission_df = pd.read_csv('/kaggle/input/playground-series-s5e12/sample_submission.csv')
```

```python
encode = LabelEncoder()
```

```python
cat_col = train_d.select_dtypes(include = ["object"]).columns
for col in cat_col:
    train_d[col] = encode.fit_transform(train_d[col])
    test_d[col] = encode.fit_transform(test_d[col])
```

```python
train_d.head()
```

```python
x = train_d.drop(['id','diagnosed_diabetes'],axis = 1)
y = train_d['diagnosed_diabetes']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = RandomForestClassifier(
    n_estimators = 200,
    max_depth = 10,
    random_state = 42
)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
x_test_submission = test_d.drop(['id'],axis = 1)
test_prediction = model.predict(x_test_submission)
submission_df['diagnosed_diabetes'] = test_prediction
submission_df.to_csv('submission_data.csv',index = False)
```
